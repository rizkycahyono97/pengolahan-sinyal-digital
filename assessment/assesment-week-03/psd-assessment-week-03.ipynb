### Import Library yang dibutuhkan

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bagian 1: Sinyal 1D

```python
# Membuat sumbu waktu dari t=0 hingga t=1 detik, dengan 500 titik data
t = np.linspace(0, 1, 500)

# Membuat Sinyal 1: Sinus dengan frekuensi 3 Hz
frekuensi1 = 3
amplitudo1 = 1.0
sinyal_1 = amplitudo1 * np.sin(2 * np.pi * frekuensi1 * t)

# Membuat Sinyal 2: Kosinus dengan frekuensi 5 Hz
frekuensi2 = 5
amplitudo2 = 0.8
sinyal_2 = amplitudo2 * np.cos(2 * np.pi * frekuensi2 * t)

# --- Visualisasi Sinyal Awal ---
plt.figure(figsize=(12, 5))

plt.subplot(2, 1, 1) # Plot pertama
plt.plot(t, sinyal_1, label=f'Sinus {frekuensi1} Hz')
plt.title('Sinyal 1')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2) # Plot kedua
plt.plot(t, sinyal_2, label=f'Kosinus {frekuensi2} Hz', color='orange')
plt.title('Sinyal 2')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()

plt.tight_layout() # Merapikan layout plot
plt.show()
```

### Operasi Penskalaan

```python
# Faktor skala
a = 2.0

# Melakukan penskalaan pada Sinyal 1
sinyal_scaled = a * sinyal_1

# --- Visualisasi Hasil Penskalaan ---
plt.figure(figsize=(10, 5))
plt.plot(t, sinyal_1, label='Sinyal Asli')
plt.plot(t, sinyal_scaled, label=f'Sinyal Diskalakan (x {a})', linestyle='--')
plt.title('Operasi Penskalaan Amplitudo')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()
```

### Operasi Penjumlahan (Addition)

```python
# Menjumlahkan Sinyal 1 dan Sinyal 2
sinyal_sum = sinyal_1 + sinyal_2

# --- Visualisasi Hasil Penjumlahan ---
plt.figure(figsize=(10, 5))
plt.plot(t, sinyal_1, label='Sinyal 1', alpha=0.5)
plt.plot(t, sinyal_2, label='Sinyal 2', alpha=0.5)
plt.plot(t, sinyal_sum, label='Hasil Penjumlahan', color='red', linewidth=2)
plt.title('Operasi Penjumlahan Sinyal')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()
```

### Operasi Penggeseran Waktu 

```python
# Waktu pergeseran (t0)
t0 = 0.2 # detik

# Melakukan pergeseran ke kanan pada Sinyal 1
# Kita tidak mengubah array 'sinyal_1', hanya sumbu 't' saat plotting
t_shifted = t + t0

# --- Visualisasi Hasil Penggeseran ---
plt.figure(figsize=(10, 5))
plt.plot(t, sinyal_1, label='Sinyal Asli')
plt.plot(t_shifted, sinyal_1, label=f'Sinyal Digeser ke Kanan ({t0}s)', linestyle='-.')
plt.title('Operasi Penggeseran Waktu')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()
```

## Bagian 1: Sinyal 2D

### Representasi Sinyal 2D

```python
# Membuat kanvas hitam 100x100 piksel
kanvas = np.zeros((100, 100))

# Gambar 1: Kotak putih di tengah
gambar_1 = kanvas.copy()
gambar_1[30:70, 30:70] = 1.0 # Nilai 1.0 merepresentasikan putih

# Gambar 2: Tanda plus (+) putih
gambar_2 = kanvas.copy()
gambar_2[45:55, 20:80] = 1.0 # Garis horizontal
gambar_2[20:80, 45:55] = 1.0 # Garis vertikal

# --- Visualisasi Sinyal 2D ---
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gambar_1, cmap='gray')
plt.title('Gambar 1: Kotak')

plt.subplot(1, 2, 2)
plt.imshow(gambar_2, cmap='gray')
plt.title('Gambar 2: Tanda Plus')

plt.show()
```

### Operasi pada Sinyal 2D

```python
# Operasi Penskalaan (mengubah kecerahan 'Gambar 1')
faktor_skala_2d = 0.6
gambar_scaled = gambar_1 * faktor_skala_2d

# Operasi Penjumlahan ('Gambar 1' + 'Gambar 2')
gambar_sum = gambar_1 + gambar_2
# Gunakan np.clip untuk memastikan nilai piksel tidak melebihi 1.0 (putih maksimal)
gambar_sum = np.clip(gambar_sum, 0, 1.0)


# --- Visualisasi Hasil Operasi 2D ---
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(gambar_1, cmap='gray')
plt.title('Gambar Asli')

plt.subplot(1, 3, 2)
plt.imshow(gambar_scaled, cmap='gray', vmin=0, vmax=1)
plt.title(f'Hasil Penskalaan (x {faktor_skala_2d})')

plt.subplot(1, 3, 3)
plt.imshow(gambar_sum, cmap='gray')
plt.title('Hasil Penjumlahan')

plt.show()
```
